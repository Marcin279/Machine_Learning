```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
max_len = iris['petal length (cm)'].max()
idx_max_len = iris['petal length (cm)'].idxmin()
min_len = iris['petal length (cm)'].min()
idx_min_len = iris['petal length (cm)'].idxmax()
print(f'Maksymalna dlugosc kwiatka {max_len} o indeksie {idx_max_len}')
print(f'Minimalna długość kwiatka {min_len} o indeksie {idx_min_len}')
```
